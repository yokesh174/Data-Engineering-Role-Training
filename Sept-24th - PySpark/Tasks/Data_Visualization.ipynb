# Sales Data Overview

This dataset contains monthly **Revenue** and **Profit** figures across different **Regions** for a year.

- **Columns**:
  - `Month`: Month of the year
  - `Revenue`: Total revenue in USD
  - `Profit`: Profit in USD
  - `Region`: Geographical region

Below is a preview of the dataset:


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct",
              "Nov", "Dec"],
    "Revenue": [12000, 15000, 17000, 14000, 18000, 22000, 24000, 20000, 23000, 25000,
                27000, 30000],
    "Profit": [3000, 3500, 4000, 3200, 4500, 6000, 6500, 5800, 6200, 7000, 7500, 8000],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West",
               "North", "South", "East", "West"]
}

df = pd.DataFrame(data)
print(df.head())
```

# Line Chart â†’ Monthly Revenue Trend

The line chart below shows the **Revenue** for each month.  
It helps visualize how revenue changes over the year.


```python
plt.plot(df["Month"], df["Revenue"], marker='o', color='blue')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

# Bar Chart â†’ Monthly Profit Comparison

The bar chart below displays the **Profit** for each month,  
highlighting months with higher and lower profits.


```python
plt.bar(df["Month"], df["Profit"], color=sns.color_palette("husl", 12))
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

# Grouped Bar Chart â†’ Revenue vs Profit per Month

This grouped bar chart compares **Revenue** and **Profit** side by side for each month,  
making it easy to see the relationship between the two metrics.


```python
import numpy as np

x = np.arange(len(df["Month"]))
width = 0.35

plt.bar(x - width/2, df["Revenue"], width, label='Revenue', color='blue')
plt.bar(x + width/2, df["Profit"], width, label='Profit', color='green')
plt.xticks(x, df["Month"])
plt.title("Revenue vs Profit per Month")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.show()
```

# Pie Chart â†’ Revenue Contribution by Region

The pie chart below shows the **percentage contribution of each Region** to the total revenue.  
It helps identify which regions generate the most revenue.


```python
region_revenue = df.groupby("Region")["Revenue"].sum()
plt.pie(region_revenue, labels=region_revenue.index, autopct='%1.1f%%', startangle=90)
plt.title("Revenue Contribution by Region")
plt.show()
```

# Scatter Plot â†’ Revenue vs Profit Correlation

The scatter plot below shows the relationship between **Revenue** and **Profit**.  
It helps visualize how profit changes as revenue increases.


```python
plt.scatter(df["Revenue"], df["Profit"], color="green")
plt.title("Revenue vs Profit Correlation")
plt.xlabel("Revenue")
plt.ylabel("Profit")
plt.show()
```

#Histogram â†’ Revenue Distribution

The histogram below shows the **distribution of monthly revenue**.  
It helps identify revenue ranges with higher frequency.


```python
plt.hist(df["Revenue"], bins=6, color="skyblue", edgecolor="black")
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

# Box Plot â†’ Profit Distribution by Region

The boxplot below shows the **distribution of Profit** across different **Regions**.  
It helps identify regional variations and potential outliers.


```python
sns.boxplot(x="Region", y="Profit", data=df, palette="Set2")
plt.title("Profit Distribution by Region")
plt.show()
```
