### Create a Simple Sales Report using Pandas
We can use **Pandas** to create and display tabular data easily:

1. Define the data as a **dictionary**, where keys are column names and values are lists of column data.  
   - Example columns: `Salesperson`, `Units Sold`, `Revenue`.  

2. Convert the dictionary into a **Pandas DataFrame** using `pd.DataFrame(data)`.  

3. Use `display(df)` to neatly show the DataFrame in tabular form (works well in Colab).


```python
import pandas as pd

# Create a simple sales report DataFrame
data = {
    'Salesperson': ['Yokesh', 'Babu', 'Jaya'],
    'Units Sold': [130, 160, 155],
    'Revenue': [40600, 60600, 55500]
}

df = pd.DataFrame(data)
display(df)
```

### Visualize Sales Revenue using Matplotlib
We can create a **bar chart** to visualize the revenue of each salesperson:

1. **Prepare the data**  
   - `salespersons` → Names on the X-axis.  
   - `revenue` → Values on the Y-axis.  

2. **Create the bar chart** using `plt.bar()`  
   - `color` → Set individual colors for each bar.  
   - `edgecolor` → Color of the bar borders.  

3. Add chart **title** and **axis labels** using `plt.title()`, `plt.xlabel()`, and `plt.ylabel()`.  

4. Use `plt.show()` to display the plot.


```python
import matplotlib.pyplot as plt

# Data from previous DataFrame
salespersons = ['Yokesh', 'Babu', 'Jaya']
revenue = [40600, 60600, 55500]

plt.bar(salespersons, revenue, color=['Yellow', 'Black', 'purple'],edgecolor= 'black')
plt.title('Sales Revenue')
plt.xlabel('Salesperson')
plt.ylabel('Revenue')
plt.show()
```
